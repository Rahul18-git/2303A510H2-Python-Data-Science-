```python
# Step 1: Imports and recursion limit
import sys
import time
import pandas as pd
from IPython.display import display

# Increase recursion limit for large inputs
sys.setrecursionlimit(2000)

# Step 2: Define recursive factorial function
def factorial_recursive(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    return 1 if n == 0 or n == 1 else n * factorial_recursive(n - 1)

# Step 3: Define iterative factorial function
def factorial_iterative(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Step 4: Run tests for various input sizes
input_numbers = [5, 10, 20, 50, 100, 200, 500, 1000]
results = []

for number in input_numbers:
    # Time recursive function
    try:
        start_time_recursive = time.time()
        factorial_recursive(number)
        end_time_recursive = time.time()
        time_recursive = end_time_recursive - start_time_recursive
    except RecursionError:
        time_recursive = float('inf')  # Recursion limit exceeded

    # Time iterative function
    start_time_iterative = time.time()
    factorial_iterative(number)
    end_time_iterative = time.time()
    time_iterative = end_time_iterative - start_time_iterative

    # Store result
    results.append({
        "input": number,
        "recursive_time": time_recursive,
        "iterative_time": time_iterative
    })

# Step 5: Create DataFrame
df_results = pd.DataFrame(results)

# Step 6: Annotate memory usage
df_results['memory_usage'] = df_results.apply(
    lambda row: 'High (Call Stack)' if row['recursive_time'] != float('inf') else 'N/A (Recursion Error)',
    axis=1
)
df_results['memory_usage_iterative'] = 'Low (Constant)'

# Step 7: Display result in Colab
display(df_results)

```
